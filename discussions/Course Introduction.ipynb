# Why This Course?

## Asset Pricing Models

Asset pricing models you have used rely on
* LOOP
* Free Portfolio Formation
* No-arbitrage

There are two widely-used approaches:
* replication (mathematical)
* equilibrium models (statistical)

**What are examples of each?**


Replication pricing relies on 
* the asset of interest being a "derivative" of the other available assets.
* liquid markets with low frictions that enable replicating trades.

Equilibrium pricing relies on
* a diversified portfolio not subject to substantial idiosyncratic risk
* an asset that is or can be transformed to be statistically stationary across a significant sample

## Need Other Methods for Pricing Idiosyncratic Assets

We will make use of
* LOOP
* Free portfolio formation
* No-arbitrage

But we will make limited use of them and explicitly account for market frictions.

The pricing model we use will be 
* **fundamental**
rather than a replicating or equilibrium model.

### Beyond classic assumptions
We will add frictions to LOOP to deal with
* tax implications of debt financing
* strategic value
* costs of financial distress
* management's incentives

We will add frictions to Free-portfolio-formation to deal with
* idiosyncratic risk
* informational asymmetry
* private equity and debt
* illiquidity
* regulatory issues

These assets are very much "incomplete" markets, making them hard to hedge or replicate.
* real options and path-dependent investment opportunities
* mergers and acquisitions

## Corporate Valuation and Quants

### Not convenient for quant methods

This approach does not easily lend itself to advanced quant methods.
* Not because it is so easy--but because it is so hard!

**What about these problems make them hard to apply advanced quant tools?**

### How can quant tools help?

Consider other developments in quant finance:
* marginal improvement making a big difference: CAPM vs Multifactor models
* combining noisy estimates: momentum strategies

Probably not going to dominate the pricing/trading of the asset class, as with options.

### Opportunities

* Improved long-short equity analysis
* Improved debt analysis
* Event strategies (IPOs, M&A, news)
* "Quantamentals" strategies
* Data-driven distressed investing
* Data-driven Private Equity and Venture Capital

## So Why This Course?

* Learn models necessary for pricing idiosyncratic assets and risks.
* Fortify knowledge in corporate finance and financial strategy.
* Get a foundation for tools that may be at the cutting edge of quant finance applications.
